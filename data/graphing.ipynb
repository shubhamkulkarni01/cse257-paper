```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
palette = {'DQN': 'b', 'SoftDQN': 'g', 'M-DQN': 'r'}
```

# Cartpole

```python
prefix = 'CartPole-v1/'
dqn = pd.read_csv(prefix + 'dqn.csv')['Value']
softdqn = pd.read_csv(prefix + 'softdqn.csv')['Value']
m_dqn_1 = pd.read_csv(prefix + 'm-dqn-1.csv')['Value']
m_dqn_2 = pd.read_csv(prefix + 'm-dqn-2.csv')['Value']
```

```python
fig = plt.figure(figsize=(10,4))

ax = fig.gca()
sns.lineplot(x = np.linspace(0, 1e5, len(dqn)), y = dqn, label='DQN', ax = ax, color=palette['DQN'])
sns.lineplot(x = np.linspace(0, 1e5, len(softdqn)), y = softdqn, label= 'SoftDQN', ax = ax, color=palette['SoftDQN'])
sns.lineplot(x = np.linspace(0, 1e5, len(m_dqn_2)), y = m_dqn_2, label= 'M-DQN', ax = ax, color=palette['M-DQN'])

sns.despine()

ax.set_title('Average reward per agent, training, CartPole', fontsize=20)
ax.set_xlabel('Step', fontsize=15)
ax.set_ylabel('Reward', fontsize=15)



ax.legend(fontsize=15)

plt.savefig('images/'+prefix+'train.png')
```

```python
frac = 0.2

part_len = lambda x: int(len(x)*frac)

fig = plt.figure(figsize=(10,4))

ax = fig.gca()
sns.lineplot(x = np.linspace(1e5 - int(1e5*frac), 1e5, part_len(dqn)), y = dqn[-part_len(dqn):], label='DQN', ax = ax, color=palette['DQN'])
sns.lineplot(x = np.linspace(1e5 - int(1e5*frac), 1e5, part_len(softdqn)), y = softdqn[-part_len(softdqn):], label= 'SoftDQN', ax = ax, color=palette['SoftDQN'])
sns.lineplot(x = np.linspace(1e5 - int(1e5*frac), 1e5, part_len(m_dqn_2)), y = m_dqn_2[-part_len(m_dqn_2):], label= 'M-DQN', ax = ax, color=palette['M-DQN'])

sns.despine()

ax.set_title(f'Average reward per agent, last {int(frac*1e2)}k steps', fontsize=20)
ax.set_xlabel('Step', fontsize=15)
ax.set_ylabel('Reward', fontsize=15)


ax.legend(fontsize=15)

```

```python
prefix = 'CartPole-v1/'
eval_dqn = pd.read_csv(prefix + 'eval-dqn.csv')['0']
eval_softdqn = pd.read_csv(prefix + 'eval-softdqn.csv')['0']
eval_mdqn = pd.read_csv(prefix + 'eval-mdqn.csv')['0']

fig = plt.figure(figsize=(10,4))

ax = fig.gca()
sns.histplot(data=eval_dqn, label='DQN', ax = ax, color='b', element='step', stat='probability')
sns.histplot(data=eval_softdqn, label='SoftDQN', ax = ax, color='g', element='step', stat='probability')
sns.histplot(data=eval_mdqn, label='M-DQN', ax = ax, color='r', element='step', stat='probability')

sns.despine()

ax.set_title('Average reward per agent, evaluation, CartPole', fontsize=20)
ax.set_xlabel('Reward', fontsize=15)
ax.set_ylabel('Frequency', fontsize=15)


ax.legend(fontsize=15)

plt.savefig('images/'+prefix+'eval.png')
```

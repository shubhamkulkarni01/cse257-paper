```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Cartpole

```python
prefix = 'cartpole/'
dqn = pd.read_csv(prefix + 'dqn.csv')['Value']
softdqn = pd.read_csv(prefix + 'softdqn.csv')['Value']
m_dqn_1 = pd.read_csv(prefix + 'm-dqn-1.csv')['Value']
m_dqn_2 = pd.read_csv(prefix + 'm-dqn-2.csv')['Value']
```

```python
fig = plt.figure(figsize=(10,4))

ax = fig.gca()
sns.lineplot(x = np.linspace(0, 1e5, len(dqn)), y = dqn, label='DQN', ax = ax)
sns.lineplot(x = np.linspace(0, 1e5, len(softdqn)), y = softdqn, label= 'SoftDQN', ax = ax)
sns.lineplot(x = np.linspace(0, 1e5, len(m_dqn_1)), y = m_dqn_1, label= 'M_DQN_1', ax = ax)
sns.lineplot(x = np.linspace(0, 1e5, len(m_dqn_2)), y = m_dqn_2, label= 'M_DQN_2', ax = ax)

sns.despine()

ax.set_title('Average reward per agent, training', fontsize=20)
ax.set_xlabel('Step', fontsize=15)
ax.set_ylabel('Reward', fontsize=15)



ax.legend(fontsize=15)
```

```python
frac = 0.2

part_len = lambda x: int(len(x)*frac)

fig = plt.figure(figsize=(10,4))

ax = fig.gca()
sns.lineplot(x = np.linspace(1e5 - int(1e5*frac), 1e5, part_len(dqn)), y = dqn[-part_len(dqn):], label='DQN', ax = ax)
sns.lineplot(x = np.linspace(1e5 - int(1e5*frac), 1e5, part_len(softdqn)), y = softdqn[-part_len(softdqn):], label= 'SoftDQN', ax = ax)
sns.lineplot(x = np.linspace(1e5 - int(1e5*frac), 1e5, part_len(m_dqn_1)), y = m_dqn_1[-part_len(m_dqn_1):], label= 'M_DQN_1', ax = ax)
sns.lineplot(x = np.linspace(1e5 - int(1e5*frac), 1e5, part_len(m_dqn_2)), y = m_dqn_2[-part_len(m_dqn_2):], label= 'M_DQN_2', ax = ax)

sns.despine()

ax.set_title(f'Average reward per agent, last {int(frac*1e2)}k steps', fontsize=20)
ax.set_xlabel('Step', fontsize=15)
ax.set_ylabel('Reward', fontsize=15)


ax.legend(fontsize=15)
```
